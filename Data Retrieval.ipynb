```python
#writing an appropriate function
results=sp.playlist_tracks(playlist_URI)
tracks= [x["track"]["uri"] for x in results["items"]]
tracks=[]
names=[]
artists=[]
artist_names=[]
artist_pop_l=[]
artist_genres_l=[]
album_l=[]
track_pop_l=[]
release_dates=[]


grabber(results,tracks,names,artists,artist_names,artist_pop_l,artist_genres_l,album_l,track_pop_l,release_dates)

```

```python
def grabber(results,tracks,names,artists,artist_names,artist_pop_l,artist_genres_l,album_l,track_pop_l,release):   
    for track in results["items"]:
        #URI
        track_uri = track["track"]["uri"]
        tracks.append(track_uri)

        #Track name
        track_name = track["track"]["name"]
        names.append(track_name)

        #Main Artist
        artist_uri = track["track"]["artists"][0]["uri"]
        artist_info = sp.artist(artist_uri)
        artists.append(artist_info)

        #Name, popularity, genre
        artist_name = track["track"]["artists"][0]["name"]
        artist_pop = artist_info["popularity"]
        artist_genres = artist_info["genres"]
        artist_names.append(artist_name)
        artist_pop_l.append(artist_pop)
        artist_genres_l.append(artist_genres)
        
        #release date
        release_date=track["track"]['album']['release_date']
        release_dates.append(release_date)

        #Album
        album = track["track"]["album"]["name"]
        album_l.append(album)

        #Popularity of the track
        track_pop = track["track"]["popularity"]
        track_pop_l.append(track_pop)
```

```python
while results['next']:
    results=sp.next(results)
    tracks_i=[]
    names_i=[]
    artists_i=[]
    artist_names_i=[]
    artist_pop_l_i=[]
    artist_genres_l_i=[]
    album_l_i=[]
    track_pop_l_i=[]
    release_dates_i=[]
    grabber(results,tracks_i,names_i,artists_i,artist_names_i,artist_pop_l_i,artist_genres_l_i,album_l_i,track_pop_l_i,release_dates)
    #extending lists
    tracks.extend(tracks_i)
    names.extend(names_i)
    artists.extend(artists_i)
    artist_names.extend(artist_names_i)
    artist_pop_l.extend(artist_pop_l_i)
    artist_genres_l.extend(artist_genres_l_i)
    album_l.extend(album_l_i)
    track_pop_l.extend(track_pop_l_i)
    release_dates.extend(release_dates_i)
    
    
```

```python
track_info=pd.DataFrame({'Tracks':tracks,'Track Name':names,'Artist': artists,
                        'Artist Name':artist_names,'Artist Popularity':artist_pop_l,
                        'Artist Genre':artist_genres_l,'Album':album_l,'Track Popularity':track_pop_l, 
                         'Release Dates':release_dates})
```
